##**YOLOV8N + MOBILENETV3 SMALL**


```python
# TAHAP 1: CUSTOM LAYER REGISTRATION (LAYER ADAPTER)
class Select_Explicit_Fix(nn.Module):
    def __init__(self, c1, c2, index=0, **kwargs):
        super().__init__()
        self.index = index

    def forward(self, x):
        if isinstance(x, list):
            return x[self.index]
        return x

# Override modul agar YOLO mengenali selector custom
conv_modules.GhostConv = Select_Explicit_Fix
tasks.GhostConv = Select_Explicit_Fix

# TAHAP 2: DEFINISI BACKBONE (MOBILENET V3 SMALL)
MODEL_NAME = 'mobilenetv3_small_100'
CORRECT_INDICES = [2, 3, 4]

class MobileNetV3Backbone(nn.Module):
    def __init__(self, *args, **kwargs):
        super().__init__()
        self.model = timm.create_model(
            MODEL_NAME, features_only=True, out_indices=CORRECT_INDICES, pretrained=True, in_chans=3
        )
    def forward(self, x):
        return self.model(x)

tasks.MobileNetV3Backbone = MobileNetV3Backbone

# Ambil dimensi channel asli dari MobileNet secara otomatis
tmp_model = timm.create_model(MODEL_NAME, features_only=True, out_indices=CORRECT_INDICES, pretrained=False)
tmp_out = tmp_model(torch.randn(1, 3, 640, 640))
REAL_CHANNELS = [x.shape[1] for x in tmp_out]
print(f"Dimensi Channel Backbone: {REAL_CHANNELS}")

# TAHAP 3: KONFIGURASI ARSITEKTUR YAML
yaml_filename = 'yolov8n_mbv3.yaml'

config_content = {
    'nc': 4,
    'scales': {'n': [0.33, 1.00, 1024]},
    'backbone': [
        [-1, 1, 'MobileNetV3Backbone', []],
        [0, 1, 'GhostConv', [REAL_CHANNELS[0], 0]], # P3
        [0, 1, 'GhostConv', [REAL_CHANNELS[1], 1]], # P4
        [0, 1, 'GhostConv', [REAL_CHANNELS[2], 2]], # P5
        [-1, 1, 'SPPF', [256, 5]]
    ],
    'head': [
        [-1, 1, 'nn.Upsample', [None, 2, 'nearest']],
        [[-1, 2], 1, 'Concat', [1]],
        [-1, 3, 'C2f', [64]],

        [-1, 1, 'nn.Upsample', [None, 2, 'nearest']],
        [[-1, 1], 1, 'Concat', [1]],
        [-1, 3, 'C2f', [64]],

        [-1, 1, 'Conv', [64, 3, 2]],
        [[-1, 7], 1, 'Concat', [1]],
        [-1, 3, 'C2f', [128]],

        [-1, 1, 'Conv', [128, 3, 2]],
        [[-1, 4], 1, 'Concat', [1]],
        [-1, 3, 'C2f', [256]],

        [[10, 13, 16], 1, 'Detect', ['nc']]
    ]
}

with open(yaml_filename, 'w') as f:
    yaml.dump(config_content, f, sort_keys=False)

# TAHAP 4: INISIALISASI & TRAINING
print(f"\n--- Memulai Training YOLOv8n + MobileNetV3 ---")
model = YOLO(yaml_filename)

model.train(
    data='/content/drive/MyDrive/RDD2022_JP_YOLO/rdd_data_jp_local.yaml',
    epochs=50,
    imgsz=640,
    batch=16,
    name='yolov8n_mbv3',
    project='yolo_comparison_test'
)
```
